## Converting Zip Codes to Block Areas Example 

#### Language: Python


```python
#Package Imports

import pandas as pd
import numpy as np
```

```python
# Make mock dataset with zip codes

a = ['Southfield', 'MA', '01259']
b = ['Stockbridge', 'MA', '01262']
c = ["Windsor", "MA", "01270"]

zipcodes = pd.DataFrame([a, b, c], columns=['town', 'state', 'zip'])

zipcodes.head()
```

```python
# Import crosswalk

crosswalk = pd.read_csv("https://www2.census.gov/geo/docs/maps-data/data/rel2020/zcta520/tab20_zcta520_tabblock20_natl.txt", sep="|")

crosswalk.head()
```

```python
# Get zip code from the 'NAMELSAD_ZCTA5_20' column by deleting the beginning word
crosswalk.NAMELSAD_ZCTA5_20 = crosswalk.NAMELSAD_ZCTA5_20.str.replace("ZCTA5 ", "")

# Pad the zipcode to ensure it is 5 digits long 
crosswalk['NAMELSAD_ZCTA5_20'] = crosswalk['NAMELSAD_ZCTA5_20'].str.pad(width=5, side='left', fillchar='0')
```

```python
# Merge datasets on zip code column
merged = pd.merge(zipcodes, crosswalk, left_on='zip', right_on='NAMELSAD_ZCTA5_20')

# Subset dataset to relevant columns
final = merged[['town', 'state', 'zip', 'OID_TABBLOCK_20', 'GEOID_TABBLOCK_20', "NAMELSAD_TABBLOCK_20"]]

final.sample(10)
```
